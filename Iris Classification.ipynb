```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as mpl
import numpy as np
%matplotlib inline
iris_dataset = load_iris()

```

```python
#Training the Algorithm

#x_train and x_test contains the features available 
#(sepal length, sepal width, petal length, petal width)

#y_train and y_test contains the iris types

x_train,x_test,y_train,y_test = train_test_split(iris_dataset["data"],
                                iris_dataset["target"],train_size=0.60,
                                test_size=0.40,random_state=1)

        
```

```python
#Visualization of train dataset
## The plots shown that data appear well distributed, good work :)

fig, ax = mpl.subplots(3,3,figsize=(12,12))

for i in range(3):
    for j in range(3):
        ax[i,j].scatter(x_train[:,j],x_train[:,i+1],c=y_train,s=60)
        
        if i == 2:
            ax[i,j].set_xlabel(iris_dataset['feature_names'][j])
        if j == 0:
            ax[i,j].set_ylabel(iris_dataset['feature_names'][i+1])
        if j > i:
            ax[i,j].set_visible(False)
```

```python
#Creating the model

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x_train,y_train)

#Testing the model
x_new = np.array([[5,2.9,1,0.2]])
prediction = model.predict(x_new)
prediction
print("Predition Result:",iris_dataset["target_names"][prediction][0])

# Get accuracy of the model
accuracy = model.score(x_test,y_test)
print("Accuracy: %.2f %%"%(accuracy*100))

```
